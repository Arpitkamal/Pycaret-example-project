# Importing the Important Libraries

```python
from pycaret.classification import *
import pandas as pd
import numpy as np
import os
```

```python
os.listdir()
```

# Loading the data

```python
pwd = os.getcwd()
test_filepath = os.path.join(pwd, "test.csv")
train_filepath = os.path.join(pwd, "train.csv")
gender_filepath = os.path.join(pwd, "gender_submission.csv")
```

```python

test_data = pd.read_csv(test_filepath)
train_data = pd.read_csv(train_filepath)
gender_data = pd.read_csv(gender_filepath)
```

```python
train_data
```

# Preprocessing 

```python
clf1 = setup(data=train_data,
             target="Survived",
             numeric_imputation='mean',
             categorical_features= ['Sex','Embarked'],
             ignore_features= ['Name','Ticket','Cabin'],
             silent=True)
```

# Comparing all the model

```python
compare_models()
```

# Creating the light Gradient Boosting Machine 

```python
lgbm_model = create_model('lightgbm')
```

# Fine tuning the model 

```python
fine_tune_lgbm_model = tune_model(lgbm_model)
```

# learning curve

```python
plot_model(estimator=fine_tune_lgbm_model, plot='learning')
```

# AUC curve

```python
plot_model(estimator = fine_tune_lgbm_model, plot = 'auc')
```

# Confusion Matrix

```python
plot_model(estimator = fine_tune_lgbm_model, plot = 'confusion_matrix')
```

# Important Feature

```python
plot_model(estimator = fine_tune_lgbm_model, plot = 'feature')
```

# From the plot we can see that `Fare`, `Age`, `Sex_female`, `Pclass_3`, `SibSp_0`, `Parch_0`, `Pclass_1` and `Embarked_S` are the Important feature

# Evaluate the model 

```python
evaluate_model(fine_tune_lgbm_model)
```

```python
#interpret_model(fine_tune_lgbm_model)
#not working 
```

```python
predict_model(fine_tune_lgbm_model, data=test_data)
```

```python
predictions = predict_model(fine_tune_lgbm_model, data=test_data)
predictions.head()
```

```python
gender_data['Survived'] = round(predictions['Score']).astype(int)
gender_data.to_csv('submission.csv',index=False)
gender_data.head()
```
